# Gaussian Plume Dispersion

Gaussian model is a common tool for modeling the plume dispersion specifically for elevated receptors.

## Unwarpped Plume Centerline using Briggs (1984)

Briggs Equation is as follows:

```python
Image(filename="./images/Briggs.png", width=400)
#%%html
#<img src="./images/Briggs.png", width=400>
```

```python
D0 = 0.63 #ft # stack dia
H0 = 40 #ft # stack height 
w0 = 50 #fps # stack exit velocity
u = 50 #fps # wind speed
Q = 15 #cfm # flow rate

x_min = 0
x_max = 500
z_min = -5
z_max = 5

xr = np.arange(x_min, x_max, 0.1)
vr = np.arange(1, u, 5)

H = np.zeros((len(xr), len(vr)))


fig, ax = plt.subplots(1)
for j, v in enumerate(vr):
    m = w0/v
    for i, x in enumerate(xr):
        H[i, j] = H0 + (3/4) ** (1/3) * (m / (1/3 + 1/m)) ** (2/3) * D0 ** (2/3) * x **(1/3)
        lb = "w0/v =" + str(round(m, 2))
    ax.plot(xr, H[:, j], label=lb)
    ax.legend(loc='best')

#print (H)
    
```

```python
xr = np.arange(x_min, x_max, 0.1)
zr = np.arange(z_min, z_max, 0.1)

vr = np.arange(1, u, 5)

fig, ax = plt.subplots(1)
for j, v in enumerate(vr):
    m = w0/v
    for i, x in enumerate(xr):
        H[i, j] = H0 + (3/4) ** (1/3) * (m / (1/3 + 1/m)) ** (2/3) * D0 ** (2/3) * x **(1/3)
        
        for z, k in enumerate(zr):
            C[i, j] = 10 ** 6 * Q / 2 / pi() / sigy / sigz / v * (exp(-0.5 * ((z - H) / sigz ) ** 2) + exp(-0.5 * ((z + H) / sigz ) ** 2))
        lb = "w0/v =" + str(round(m, 2))
    ax.plot(xr, H[:, j], label=lb)
    ax.legend(loc='best')
```
